## Библиотека numpy

На массивах numpy строятся датафреймы - это тип данных (наши таблицы, но об этом позже)

Что интересного есть у numpy и что нам пригодится? 

Массив numpy напоминает собой список. И обычные списки мы можем переводить в массивы numpy и выполнять различные операции. 

Например, у нас есть список x=  [1,2,3,4], чтобы в списке умножить каждый элемент на 2, в питоне нам бы пришлось делать цикл for и тд. В numpy мы можем просто сделать x*2. Так работает с любой математической операцией. 

Это пригодится нам для решения некоторых задач уровня B. 

Чтобы создать массив numpy, пользуемся функцией array



```python
x = [1,2,3,4]
x*2
```

```python
x = np.array([1,2,3,4])
x*2
```

```python
import numpy as np
x = np.array([1,2,3,4])
y = np.array([3,4,5,6])
print((x-y)**2)
```

```python
+
-
* - умножение 
** - степень (4**2 = 16)
/ - деление 
```

### Фильтрация 

Когда мы хотим отсмотреть определенные данные в таблице, например, сколько людей старше 60 лет и т.д.

```python
Для фильтрации нужно поставить условие. Условия как и в питоне.
1. >
2. <
3. >=
4. <=
5. ==
6. !=
```

Можно ставить не только простые условия, но и сложные. Например, старше 30 лет, но младше 50. Для этого нужны операнды.

| - операнд "или" (если хотя бы выполняется одно из условий)

& - операнд "и" (должны выполняться оба условия)

```python
# первый способ
data['Age'] > 40
```

```python
(data['Age'] > 40).sum()
```

```python
# второй способ
data[data['Age'] > 40]
```

```python
data[data['Age'] > 40]['SystolicBP']
```

```python
# отфильтрованную таблицу можно присвоить другой переменной 
df = data[data['Age'] > 40]
```

```python
df['SystolicBP'].max()
```

```python
(data['Age'] > 30)&(data['Age'] < 50)
```

```python
# со сложным условием
data[(data['Age'] > 30) & (data['Age'] < 50)]
```

```python
data[(data['RiskLevel'] == 'high risk') | (data['RiskLevel'] == 'mid risk')]
```

Интересный лайфхак, который пригодится на экзамене. Допустим, нам нужно найти максимальный возраст пациентов, у которого уровень сахара равен 6. 

```python
# Двойная индексация
data[data['BS'] == 6]['Age'].max()
```

### Типы данных (переменных)

1. Количественные:
- дискретные (количество детей, год)
- непрерывные (возраст)
2. Качественные(категориальные):
- порядковые/ранговая(степени ожога)
- номинальные (цвет):
    - бинарные (пол, true/false)

1. Age: возраст, когда пациентка забеременела 
2. SystolicBP: верхнее значение артериального давления в мм рт. ст. 
3. DiastolicBP: нижнее значение артериального давления в мм рт. ст. 
4. BS: уровень глюкозы в крови в ммоль/л - 
5. HeartRate: нормальная частота сердечных сокращений в состоянии покоя в ударах в минуту - 
6. BodyTemp - температура тела по фаренгейту - 
7. Risk Level: прогнозируемый риск здоровью (1 - высокий, 2 - средний, 3 - низкий)
